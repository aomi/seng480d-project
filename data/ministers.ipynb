```python
import pandas as pd
```

```python
def add_metadata(d: pd.DataFrame, i):
    d['parliament'] = i
    return d

df = pd.concat([add_metadata(pd.read_csv(f"mps/mp_{i}.csv"), i) for i in range(36, 45)])
```

```python
df['name'] = df['Last Name'] + ',' + df['First Name']
```

```python
df['Political Affiliation'].unique()
```

```python
d = df['name'].unique()
d.sort()
d
```

```python
df.rename(columns={
    'Honorific Title': 'title',
    'First Name': 'first_name',
    'Last Name': 'last_name',
    'Constituency': 'constituency',
    'Province / Territory': 'province',
    'Political Affiliation': 'party',
    'Start Date': 'start_date',
    'End Date': 'end_date'
}, inplace=True)
df.to_json('../src/ministers.json', orient='records')
```
