```python
import pandas as pd
import numpy as np
import math
```

```python
org_df = pd.read_csv('expenditure_data.csv')
org_df["Member's per diem expenses"] = org_df["Member's per diem expenses"].apply(lambda x: 0 if math.isnan(x) else x)
org_df['3-Travel'] = org_df['Member'] + org_df['Designated traveller'] + org_df['Dependants'] + org_df['Employees'] + org_df['Member\'s accommodation expenses'] + org_df['Member\'s per diem expenses'] + org_df['Member\'s secondary residence expenses']
# org_df.head()
# org_df.loc[(org_df["Year"]==2018) & (org_df["Quarter"]==4)]


```

```python
# Year,Quarter,Name,Status,Constituency name,Constituency size,Number of electors,1-Employees' salaries,2-Service Contracts,3-Travel,Member,Designated traveller,Dependants,Employees,Member's accommodation expenses,
# Member's per diem expenses,Member's secondary residence expenses,4-Hospitality,5-Gifts,6-Advertising,7-Printing,Householders,Ten percenters,Other printing related expenses,8-Offices,
# "Constituency office leases, insurance and utilities","Furniture, furnishing and equipment purchases",Equipment rentals,Informatics and telecommunication equipment purchases,
# Telecommunication services,Repairs and maintenance,Postage and courier services,Materials and supplies,Training,Total


df = pd.read_csv('MembersExpenditures.2021Q2.csv')
df.insert(0,'Year',2020)
df.insert(1,'Quarter',2)
df.rename({"Constituency": "Constituency name"}, axis=1, inplace=True)
df.drop('Caucus',inplace=True, axis=1)
# df.rename({"Constituency": "Constituency name"}, axis=1, inplace=True)
df.insert(3,'Status','')
df.insert(5,'Constituency size',0)
df.insert(6,'Number of electors',0)
df.rename({"Salaries": "1-Employees' salaries"}, axis=1, inplace=True)
df.insert(8,'2-Service Contracts',0)
df['2-Service Contracts'] = df['Contracts']
df.rename({"Travel": "3-Travel"}, axis=1, inplace=True)
df.insert(10,'Member',0)
df.insert(11,'Designated traveller',0)
df.insert(12,'Dependants',0)
df.insert(13,'Employees',0)
df.insert(14,"Member's accommodation expenses",0)
df.insert(15,"Member's per diem expenses",0)
df.insert(16,"Member's secondary residence expenses",0)
df.insert(17,'4-Hospitality',0)
df['4-Hospitality'] = df['Hospitality']
df.insert(18,'5-Gifts',0)
df.insert(19,'6-Advertising',0)
df.insert(20,'7-Printing',0)
df.insert(21,'Householders',0)
df.insert(22,'Ten percenters',0)
df.insert(23,'Other printing related expenses',0)
df.insert(24,'8-Offices',0)
df.insert(25,'Constituency office leases, insurance and utilities',0)
df.insert(26,'Furniture, furnishing and equipment purchases',0)
df.insert(27,'Equipment rentals',0)
df.insert(28,'Informatics and telecommunication equipment purchases',0)
df.insert(29,'Telecommunication services',0)
df.insert(30,'Repairs and maintenance',0)
df.insert(31,'Postage and courier services',0)
df.insert(32,'Materials and supplies',0)
df.insert(33,'Training',0)
df.insert(34,'Total',0)
df['Total'] = df["1-Employees' salaries"] + df['3-Travel'] + df["4-Hospitality"] + df["2-Service Contracts"]
df.drop(['Hospitality','Contracts'],inplace=True, axis=1)

df['Name'] = df['Name'].apply(lambda x: x.replace(' ',''))

name_list = list(org_df.loc[org_df['Year'] == 2020]["Name"])

for name in list(df["Name"]):

    if name in name_list:
        ind1 = df.index[(df["Name"]==name) & (df["Year"]==2020)][0]
        ind2 = org_df.index[(org_df["Name"]==name) & (org_df["Year"]==2020)][0]
        
        df.at[ind1,"1-Employees' salaries"] += org_df.at[ind2,"1-Employees' salaries"]
        df.at[ind1,"2-Service Contracts"] += org_df.at[ind2,"2-Service Contracts"]
        df.at[ind1,"3-Travel"] += org_df.at[ind2,"3-Travel"]
        df.at[ind1,"4-Hospitality"] += org_df.at[ind2,"4-Hospitality"]
        df.at[ind1,"Total"] += org_df.at[ind2,"Total"]
        

df.head()
# len(df.index)
# df.loc[0]["Total"]
```
